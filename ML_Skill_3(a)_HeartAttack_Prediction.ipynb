```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split

dt=pd.read_csv('HeartStroke(f)Prediction.csv')

plt.hist(dt['Age'],bins=5,color='Pink',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()

plt.hist(dt['HeartDisease'],bins=5,color='yellow',edgecolor='black')
plt.xlabel('Heart stroke')
plt.ylabel('Count')
plt.title('Heart Stroke prediction')
plt.show()

le=LabelEncoder()
le.fit(dt.Sex.drop_duplicates())
dt.Sex=le.transform(dt.Sex)

le.fit(dt.ChestPainType.drop_duplicates())
dt.ChestPainType=le.transform(dt.ChestPainType)

le.fit(dt.RestingECG.drop_duplicates())
dt.RestingECG=le.transform(dt.RestingECG)

le.fit(dt.ExerciseAngina.drop_duplicates())
dt.ExerciseAngina=le.transform(dt.ExerciseAngina)

le.fit(dt.ST_Slope.drop_duplicates())
dt.ST_Slope=le.transform(dt.ST_Slope)

x=dt.drop(columns=['HeartDisease'])
y=dt['HeartDisease']
print('X is:',x,'\n Y is:\n',y)
xtrn,xtst,ytrn,ytst=train_test_split(x,y,test_size=0.2,random_state=42)
```

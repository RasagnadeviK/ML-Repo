```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

datadict={
    'sno':[1,2,3,4,5,6],
    'Temperature':[0,20,40,60,80,100],
    'Pressure':[0.0002,0.0012,0.0060,0.0300,0.0900,0.2700]
}
dt=pd.DataFrame(datadict)
dt.to_csv("Data.csv")

x=dt.iloc[:, 1:2].values
y =dt.iloc[:, 2].values

lr=LinearRegression()
lr.fit(x,y)

pf=PolynomialFeatures(degree=2)
xpf=pf.fit_transform(x)
pf.fit(xpf,y)

lr2=LinearRegression()
lr2.fit(xpf,y)

plt.scatter(x,y,color='Green')
#plt.plot(x,lr.predict(x),color='red')
plt.title('Linear Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()

plt.scatter(x,y,color='purple')
plt.plot(x,lr2.predict(pf.fit_transform(x)),color='pink')
plt.title('Polynomial Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()

prdt=110.0
for i in range(5):
  prdtarr=np.array([[prdt]])
  print('for temperature ',prdt,'pressure is:',lr.predict(prdtarr))
  prdt+=10.0
```

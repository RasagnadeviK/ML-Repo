```python
import matplotlib.pyplot as plt
import numpy as np

sem1=[10,10,10,9,9]
sem2=[10,9,10,9,10]
sem3=[9,10,10,10,9]
sem4=[10,10,9,9,10]

act=['Sem1','Sem2','Sem3','Sem4']

slices=[9.6,9.7,9.8,9.8]
colr=['b','c','g','y']
slic=[np.mean(sem1),np.mean(sem2),np.mean(sem3),np.mean(sem4)]
col=['c','g','y','r']

plt.pie(slices,labels=act,colors=colr,startangle=90,shadow= True,explode=(0.1,0.1,0.3,0.2),autopct='%1.1f%%')
plt.title('Semester wise CGPA')
plt.show()

plt.pie(slic,labels=act,colors=col,startangle=90,shadow= True,explode=(0.1,0.1,0.3,0.2),autopct='%1.1f%%')
plt.title("Semester wise points")
plt.show()
```

```python
import numpy as np
import scipy.stats as ss


x=np.arange(10,20)

print(x)
y=np.array([10,1,2,6,7,8,3,4,5,9])
print(y)

r=np.corrcoef(x,y)
print(r)
print(r[0][0]," ",r[0][1])
print(r[1][0])
print(round(r[1][1]))

print(ss.pearsonr(x,y))

print(ss.spearmanr(x,y))

print(ss.kendalltau(x,y))
```

```python
import pandas as pd
import numpy as np
import scipy.stats as ss

data=pd.read_csv('DataID.csv')
print(data)
age=np.array(data['Age'])
inc=np.array(data['Inc'])
scr=np.array(data['Scr'])
print(data)

ra=np.corrcoef(age,inc)
ri=np.corrcoef(inc,scr)
rs=np.corrcoef(scr,age)

print("Correlation of age and income :",ra)
print("Correlation of score and income :",ri)
print("Correlation of age and score :",rs)

print("Pearson correlation of age and income :",ss.pearsonr(age,inc),"\n")
print("Spearman correlation of age and income :",ss.spearmanr(age,inc),"\n")
print("Kendalltau correlation of age and income :",ss.kendalltau(age,inc),"\n")

print("Pearson correlation of score and income :",ss.pearsonr(scr,inc),"\n")
print("Spearman correlation of score and income :",ss.spearmanr(scr,inc),"\n")
print("Kendalltau correlation of score and income :",ss.kendalltau(scr,inc),"\n")

print("Pearson correlation of age and score :",ss.pearsonr(inc,age),"\n")
print("Spearman correlation of age and score :",ss.spearmanr(inc,age),"\n")
print("Kendalltau correlation of age and score :",ss.kendalltau(inc,age),"\n")

```
